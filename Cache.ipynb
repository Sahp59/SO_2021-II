```python
"""
Created on Sat Apr  3 01:59:32 2021

@author: said
"""

import random
import sys

def representacionRam():
   L1 = [["Indice","Dato", "Tamaño"]] #Indice,Dato,Tamaño
   Row = []
   Ram = random.randrange(5,8000)
   resultado = 0

   x = 0    #Ayuda para el indice
   
   while resultado < Ram:
      dato = cadena()
      Row.append(x)                       #Indice
      Row.append(dato)                    #Dato
      Row.append(sys.getsizeof(dato))     #Tamaño
      
      L1.append(Row)                      #Agregamos las filas en la tabla
      Row = []                            #Vaciamos la fila
      resultado += sys.getsizeof(dato)    #Sumamos la memoria que hemos agregado
      
      x = x + 1                           #Esto nos da el indice de los elementos

   imprimirArray(L1)
   print("\nLA RAM TIENE COMO MAXIMO: " , Ram , "Kb \t")
   print("LA RAM TIENE: " , resultado , "Kb Almacenados")
      

def memoriaTresNiveles():
   L1 = [["Indice","Dato", "Tamaño"]]     #Memoria Ram de 128Kb
   L2 = [["Indice","Dato", "Tamaño"]]     #Memoria Ram de 1Mb
   L3 = [["Indice","Dato", "Tamaño"]]     #Memoria Ram de 50Mb
   
   Row = []    #Nos agrega todo en columnas
   resultado = 0
   x = 0    #Ayuda para el indice
   
   while resultado < 6000000:             #Este valor nos indica si se ocuparan las 3 Memorias
      dato = cadena()
      Row.append(x)                       #Indice
      Row.append(dato)                    #Dato
      Row.append(sys.getsizeof(dato))     #Tamaño
   
      resultado += sys.getsizeof(dato)    #Sumamos la memoria que hemos agregado
   
      if resultado < 128000:
         L1.append(Row)
         L2.append(Row)
         L3.append(Row)
      if resultado < 1000000 and resultado > 128000:
         L2.append(Row)
         L3.append(Row) 
      if resultado < 50000000 and resultado > 1000000:
         L3.append(Row) 
      
      x = x + 1
      Row = []   
   
   print(len(L1), "Indices de la Ram de 128kb")          #Agrego estas lineas para observar cuantos indices tiene cada Ram
   print(len(L2), "Indices de la Ram de 1Mb")          #Agrego estas lineas para observar cuantos indices tiene cada Ram
   print(len(L3), "Indices de la Ram de 50Mb")          #Agrego estas lineas para observar cuantos indices tiene cada Ram
   
   #imprimirArray(L1)      #En caso de querer ver las Tablas de Ram quite en primer gato de esta linea
   #imprimirArray(L2)      #En caso de querer ver las Tablas de Ram quite en primer gato de esta linea
   #imprimirArray(L3)      #En caso de querer ver las Tablas de Ram quite en primer gato de esta linea
   
   
def cadena():
   Dato = ""
   longitud = random.randint(1, 11)
   for i in range(0,longitud,1):
       Dato += chr(random.randint(0, 127))
   return Dato

def imprimirArray(arreglo):
   for i in range(0, len(arreglo)-1):
      print(arreglo[i])

print("Representacion de Ram\n")      
representacionRam() 
print("\nMemoria de Tres Niveles\n")  
memoriaTresNiveles()
```
