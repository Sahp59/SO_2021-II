```
import random

class Cola:
    def __init__(self):
        self.items = []

    def estaVacia(self):
        return self.items == []

    def agregar(self, item):
        self.items.insert(0, item)

    def avanzar(self):
        return self.items.pop()

    def tamano(self):
        return len(self.items)


def main():
    array = []
    Buffer = Cola()
    
    # Tabla de 1000 elementos
    for i in range(1000):
        arraySegundo = [[i], [random.uniform(11, 15)], random.randint(1, 5)]
        array.append(arraySegundo)

    for i in range(5, -1, -1):
        for j in range(len(array)):
            if array[j][2] == i and Buffer.tamano() < 800:
                Buffer.agregar(array[j])

    print("El numero de datos dentro del buffer es: ", Buffer.tamano())
    
    # Espera a que se acceda para imprimir los datos
    for i in range(Buffer.tamano()):
        print(Buffer.avanzar())
        print(" ")

    print("El numero de datos dentro del buffer es: ", Buffer.tamano())


main()
```
